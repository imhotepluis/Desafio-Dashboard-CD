```python
# IQR Distancia Inter Quartil

q1 = data_nova['age'].quantile(0.25)
q3 = data_nova['age'].quantile(0.75)

IQR = q3 - q1

print(f'IQR = {IQR}')
```

```python
idade_outliers = data_nova[(data_nova['age'] < q1 - (IQR * 1.5)) | (data_nova['age'] > q3 + (IQR * 1.5))]
```

```python
idade_outliers
```

```python
idade_inliers = data_nova[(data_nova['age'] >= q1 - (IQR * 1.5)) & (data_nova['age'] <= q3 + (IQR * 1.5))]
```
